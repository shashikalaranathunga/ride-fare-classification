```python
##Importing the dataset
dataset = pd.read_csv('train.csv')

dataset['label'] = (dataset['label'] == 'correct').astype('int')

# overwriting data after changing format 
dataset['pickup_time']= pd.to_datetime(dataset["pickup_time"]) 
dataset['drop_time']= pd.to_datetime(dataset["drop_time"]) 
```

```python
dataset['pickup_time_year'] = dataset['pickup_time'].dt.year
dataset['pickup_time_month'] =dataset['pickup_time'].dt.month
dataset['pickup_time_week'] =dataset['pickup_time'].dt.week
dataset['pickup_time_day'] = dataset['pickup_time'].dt.day
dataset['pickup_time_hour'] =dataset['pickup_time'].dt.hour
dataset['pickup_time_minute'] =dataset['pickup_time'].dt.minute
dataset['pickup_time_dayofweek'] =dataset['pickup_time'].dt.dayofweek
```

```python
dataset['drop_time_year'] = dataset['drop_time'].dt.year
dataset['drop_time_month'] =dataset['drop_time'].dt.month
dataset['drop_time_week'] =dataset['drop_time'].dt.week
dataset['drop_time_day'] = dataset['drop_time'].dt.day
dataset['drop_time_hour'] =dataset['drop_time'].dt.hour
dataset['drop_time_minute'] =dataset['drop_time'].dt.minute
dataset['drop_time_dayofweek'] =dataset['drop_time'].dt.dayofweek
```

```python
dataset.isnull().sum()
#dataset.dropna(how='any')
#print(dataset.shape)
```

```python
#Filling missing values with a test statistics like mean, median or mode.
#Showing here with mean only

mean_value=dataset['additional_fare'].median()
dataset['additional_fare']=dataset['additional_fare'].fillna(mean_value)

mean_value=dataset['duration'].median()
dataset['duration']=dataset['duration'].fillna(mean_value)


mean_value=dataset['meter_waiting'].median()
dataset['meter_waiting']=dataset['meter_waiting'].fillna(mean_value)



mean_value=dataset['meter_waiting_fare'].median()
dataset['meter_waiting_fare']=dataset['meter_waiting_fare'].fillna(mean_value)




mean_value=dataset['meter_waiting_till_pickup'].median()
dataset['meter_waiting_till_pickup']=dataset['meter_waiting_till_pickup'].fillna(mean_value)



mean_value=dataset['fare'].median()
dataset['fare']=dataset['fare'].fillna(mean_value)
print(dataset.head(5))
```

```python

##here find the label values 1 and o 

correct= dataset.loc[dataset['label'] == 1].values
incorrect = dataset.loc[dataset['label'] == 0].values
print(len(correct))
```

```python
print(correct1.shape)
print(incorrect1.shape)
# Highly imbalanced dataset (bias towards the positive class)
```

```python
# Balancing classes by undersampling the positive class
correct1 = correct1[0:incorrect1.shape[0]]
correct1_label = correct1_label[0:incorrect1.shape[0]]
correct1_label.shape
incorrect1_label.shape
```

```python
data=np.vstack((correct1,incorrect1))
#print(data.shape)


x= pd.DataFrame(data = data, columns=col2use)
x.head()

#y=np.vstack((correct1_label,incorrect1_label))
y=np.concatenate((correct1_label,incorrect1_label), axis=None)




from sklearn.utils import shuffle

x, y = shuffle(x, y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
X_train.shape, X_test.shape
```
